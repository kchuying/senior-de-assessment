```python
#Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# These two lines are to set the size of the plot/chart
## For Matplotlib
%matplotlib inline
plt.rc('figure', figsize=(15,5))
## For Seaborn
sns.set(rc= {"figure.figsize": (15, 5)})
```

```python
#Read cleansed data - Singapore covid cases
df = pd.read_csv("cleansed_covid_data.csv")
df.head()
```

```python
#Plot line chart to show trend of total confirmed covid cases in SG
df.plot(x="date", y="total_confirmed_cases", kind="line")
plt.ticklabel_format(style='plain', axis='y')
plt.title('Covid Cases in Singapore', fontsize=20)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total No. of Confirmed Cases', fontsize=14)
```

```python
#Plot line chart to show trend of daily confirmed covid cases in SG
df.plot(x="date", y="daily_confirmed", kind="line")
plt.ticklabel_format(style='plain', axis='y')
plt.title('Covid Cases in Singapore', fontsize=20)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total No. of Daily Confirmed Cases', fontsize=14)
```

```python
#Aggregate number of daily confirmed cases by year month
yearMonth = df.groupby(by = ['year', 'month']).sum()['daily_confirmed'].unstack().dropna()
yearMonth
```

```python
#Aggregate number of daily confirmed cases by day month
dayMonth = df.groupby(by = ['day', 'month']).sum()['daily_confirmed'].unstack().dropna()
dayMonth
```

```python
#Identify the hot period of cases via heatmap
plt.figure(figsize = (12,6))
sns.heatmap(yearMonth, cmap = 'viridis')
```

```python
#Identify the hot period of cases via heatmap
plt.figure(figsize = (12,6))
sns.heatmap(dayMonth, cmap = 'viridis')
```
